```python
import pandas as pd 

df= pd.read_excel("Vendas.xlsx")
df
```

```python
faturamento_final = df['Valor Final'].sum()
print("R$",faturamento_final)
```

```python
faturamento_por_loja = df[['ID Loja', 'Valor Final']].groupby("ID Loja").sum()
display(faturamento_por_loja)
```

```python
faturamento_por_produto = df[['ID Loja','Produto','Valor Final']].groupby(['ID Loja', 'Produto']).sum()
display(faturamento_por_produto)
```

```python
import matplotlib as pld

faturamento_por_produto = df.groupby(['ID Loja', 'Produto'])['Valor Final'].sum().unstack()
faturamento_por_produto.plot(kind='bar', fingsize = (10,6))

pld.title('Faturamento por produto em cada loja')
pld.xlabel('ID Loja')
pld.ylabel('Faturamento (R$)')
pld.legend('Produto', box_produto = (1.05,1), loc='upper left')

pld.tinght_layout()
pld.show
```
